```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
countries = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DS & AI - M4/world_table_country.csv')
#countries
```

```python
def plot_color_palette(palette: str):
    figure = sns.palplot(sns.color_palette())
    plt.xlabel(palette)
    plt.show(figure)
    
palettes = ['deep','muted','pastel','bright','dark','colorblind','Spectral','Paired','Set1','Set2','husl']
for palette in palettes:
    sns.set(palette=palette)
    plot_color_palette(palette)
```

```python
sns.reset_orig()
```

```python
for style in ['white','dark','whitegrid','darkgrid']:
  sns.set_style(style) 
  plt.plot()
  sns.histplot(countries['LifeExpectancy']) 
  plt.xlabel(style)
  plt.ylabel('')
  plt.show()
```
